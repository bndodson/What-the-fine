```python
class Tickets(param.Parameterized):
    year = param.ObjectSelector(default=2017, objects=list(dataset.Year.unique()))
    show_legend = param.Boolean(default=True)
    
    def get_data(self):
        df = dataset[dataset['Year']==self.year]
        return df
    
    def mean_fine(self):
        data = self.get_data()
        data = data[data['variable']=='finemean']
        data = data.rename(columns = {"value":"Mean Fine"}) 
        data['Mean Fine'] = data['Mean Fine'].round()
        title='Mean Parking Violation Fine'
        plot = data.hvplot(c='Mean Fine', 
                 geo=True, 
                 line_width=0, 
                 cmap=bmy, 
                 tiles='CartoLight', 
                 xaxis=None, 
                 yaxis=None, 
                 width=400, height=400,
                 title=title, clim=(30,100))
        return plot
    
    def median_fine(self):
        data = self.get_data()
        data = data[data['variable']=='finemedian']
        data = data.rename(columns = {"value":"Median Fine"}) 
        data['Median Fine'] = data['Median Fine'].round()
        title = 'Median Parking Violation Fine'
        plot = data.hvplot(c='Median Fine', 
                 geo=True, 
                 line_width=0, 
                 cmap=bmy, 
                 tiles='CartoLight', 
                 xaxis=None, 
                 yaxis=None, 
                 width=400, height=400,
                 title=title, clim=(20,80))
        return plot
    
    def total_violations(self):
        data = self.get_data()
        data = data[data['variable']=='finecount']
        data = data.rename(columns = {"value":"Parking Violations"}) 
        data['Parking Violations'] = data['Parking Violations'].round()
        title = 'Total Parking Violations'
        plot = data.hvplot(c='Parking Violations', 
                 geo=True, 
                 line_width=0, 
                 cmap=bmy, 
                 tiles='CartoLight', 
                 xaxis=None, 
                 yaxis=None, 
                 width=400, height=400,
                 title=title)
        return plot
    
    def violations_norm(self):
        data = self.get_data()
        data = data[data['variable']=='violations_norm']
        data = data.rename(columns = {"value":"Violations (Normalized to Area)"}) 
        title = 'Total Parking Violations (Normalized to Area)'
        plot = data.hvplot(c='Violations (Normalized to Area)', 
                 geo=True, 
                 line_width=0, 
                 cmap=bmy, 
                 tiles='CartoLight', 
                 xaxis=None, 
                 yaxis=None, 
                 width=400, height=400,
                 title=title)
        return plot
    
app = Tickets(name='')
```
